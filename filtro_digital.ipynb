No exemplo acima introduzimos o conceito de **bloco de atraso**. Note que cada saída $y[n]$ é a média das **N** amostras do sinal de entrada $x[m-k_N]$. Ou seja
$$
y3[n] = \frac{x[n-0]+x[n-1]+x[n-2]+x[n-3]+x[n-4]}{5}
$$
Podemos reescrever a equação de maneira mais geral, na forma:
$$
y3[n] = \frac{1}{5}\sum_{k=0}^{4}x[n-k]
$$

```
N = 3

y4 = np.zeros(x.shape)

for n in range(N-1, len(x)):
  for k in range(0, N):
    y4[n] += x[n-k]
  y4[n] = y4[n]/N

y5 = np.zeros(x.shape)

for n in range(N-1, len(x)):
  for k in range(0, N):
    y5[n] += x[n-k]/N
```

```
def convolve(x, h):
  ans = np.empty([]);
  aux = np.zeros(h.shape);

  for n in range(1, len(x)):
    aux = np.delete(np.insert(aux, 0, x[n]),-1)
    # print(aux)
    ans = np.hstack((ans, np.dot(aux, h.real)))

  return ans
```

```
def window(N, window='kaiser', alpha=2):
  if window == 'hanning':
    n = np.linspace(-N/2, N/2, N)
    return 0.5*(1+np.cos(2*np.pi*n/N))
  if window == 'hamming':
    a0=0.53836
    n = np.linspace(0, N, N)
    return a0-(1-a0)*(np.cos(2*np.pi*n/N))
  if window == 'gaussian':
    sigma=0.5
    n = np.linspace(0, N, N)
    return np.exp(-0.5*((n-N/2)/(sigma*N/2))**2)
  if window == 'blackmanHarris':
    a0, a1, a2, a3 = 0.35875, 0.48829, 0.14128, 0.01168
    n = np.linspace(0, N, N)
    return a0-a1*np.cos(2*np.pi*n/N)+a2*np.cos(4*np.pi*n/N)-a3*np.cos(6*np.pi*n/N)
  if window == 'flattopWindow':
    a0, a1, a2, a3, a4 = 0.21557895, 0.41663158, 0.277263158, 0.083578947, 0.006947368
    n = np.linspace(0, N, N)
    return a0-a1*np.cos(2*np.pi*n/N)+a2*np.cos(4*np.pi*n/N)-a3*np.cos(6*np.pi*n/N)+a4*np.cos(8*np.pi*n/N)
  if window == 'kaiser':
    n = np.linspace(0, N, N)
    return np.i0(np.pi*alpha*np.sqrt(1-pow((2*n/N)-1,2)))/np.i0(np.pi*alpha)
```

```
def filter2(x, b, a):
  ans = np.empty([]);
  auxn = np.zeros(b.shape);
  auxd = np.zeros(a.shape);

  for k in range(1, len(x)):
    # print(auxd)
    ansn = np.dot(auxn, b)
    ansd = np.dot(auxd, a)
    y = ansn + ansd

    ans = np.hstack((ans, y))
    
    auxn = np.delete(np.insert(auxn, 0, x[k]),-1)
    auxd = np.delete(np.insert(auxd, 0, y),-1)

  return ans
```
